```python
import csv

# base_list = "string1, string2, string3"
# comparision_list = "string2, string3, string4"

def get_common_tags(base_list, comp):

    dataLists = [base_list, comp]

    for dataList in dataLists: 

        lines = dataList.splitlines()
        # print(lines)

        reader = csv.reader(lines)
        parsed_csv = list(reader)
        # print(parsed_csv)

        if dataList == base_list:
            mydataList = parsed_csv
            baseList = mydataList[0]
            baseListPrepped = []
            for tag in baseList:
                strippedTag = tag.strip()
                baseListPrepped.append(strippedTag)
        elif dataList == comp:
            mydataList2 = parsed_csv
            compList = mydataList2[0]
            compListPrepped = []
            for tag in compList:
                strippedTag = tag.strip()
                compListPrepped.append(strippedTag)

    
    # print("prepped base", baseListPrepped)
    # print("prepped comparision", compListPrepped)

    commonTagList = []

    for baseTag in baseListPrepped:
        for compTag in compListPrepped:
            if baseTag == compTag:
                commonTagList.append(baseTag)

    textfile = open("a_file.txt", "w")
    for element in commonTagList:
        textfile.write(element + "\n")
    textfile.close()

    print("common list", commonTagList)

```
